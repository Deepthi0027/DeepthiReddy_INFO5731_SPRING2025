# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import math

def precedence(op):
    if op in ('+', '-'):
        return 1
    if op in ('*', '/', '%'):
        return 2
    if op in ('**', 'sqrt'):
        return 3
    return 0

def apply_operation(values, operators):
    op = operators.pop()
    if op == 'sqrt':
        if not values:
            raise ValueError("Invalid expression: Missing operand for sqrt")
        values.append(math.sqrt(values.pop()))
    else:
        if len(values) < 2:
            raise ValueError("Invalid expression: Insufficient values for operation")
        right = values.pop()
        left = values.pop()
        if op == '+':
            values.append(left + right)
        elif op == '-':
            values.append(left - right)
        elif op == '*':
            values.append(left * right)
        elif op == '/':
            if right == 0:
                raise ZeroDivisionError("Division by zero is not allowed.")
            values.append(left / right)
        elif op == '%':
            values.append(left % right)
        elif op == '**':
            values.append(left ** right)

def evaluate_expression(expression):
    tokens = tokenize(expression)
    values = []
    operators = []
    index = 0

    while index < len(tokens):
        token = tokens[index]
        if token.isdigit() or '.' in token:
            values.append(float(token) if '.' in token else int(token))
        elif token == 'sqrt':
            operators.append(token)
        elif token == '(':
            operators.append(token)
        elif token == ')':
            while operators and operators[-1] != '(':
                apply_operation(values, operators)
            if not operators:
                raise ValueError("Mismatched parentheses in expression")
            operators.pop()
        else:
            while (operators and precedence(operators[-1]) >= precedence(token)):
                apply_operation(values, operators)
            operators.append(token)
        index += 1

    while operators:
        apply_operation(values, operators)

    if len(values) != 1:
        raise ValueError("Invalid expression: Incorrect number of values left in stack")
    return values[0]

def tokenize(expression):
    tokens = []
    num = ''
    i = 0
    while i < len(expression):
        char = expression[i]

        if char.isdigit() or char == '.':
            num += char
        else:
            if num:
                tokens.append(num)
                num = ''

            if expression[i:i+4] == "sqrt":
                tokens.append("sqrt")
                i += 3
            elif expression[i:i+2] == "**":
                tokens.append("**")
                i += 1
            elif char in "+-*/%()":
                tokens.append(char)
        i += 1

    if num:
        tokens.append(num)

    return tokens

def main():
    print("Stack-Based Command-Line Calculator")
    print("Supported operations: +, -, , /, * (exponentiation), % (modulus), sqrt(x)")
    print("Type 'exit' to quit the calculator.")

    while True:
        expression = input("Enter expression: ")
        if expression.lower() == "exit":
            print("Exiting calculator. Goodbye!")
            break
        try:
            print("Result:", evaluate_expression(expression))
        except Exception as e:
            print("Error:", e)

main()
```
